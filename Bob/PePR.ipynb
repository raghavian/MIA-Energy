```python
import numpy as np
import plotly.express as px
```

```python
ressource = np.linspace(0, 1, 2**10)
```

```python
omega = 1
performance = np.cumsum(np.sin(2*np.pi*omega*ressource)**32)
performance /= np.max(performance)
```

```python
px.line(x=ressource, y=performance)
```

```python
pepr = performance / (1 + ressource)
```

```python
px.line(x=ressource, y=pepr)
```
